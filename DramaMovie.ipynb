```python
import pandas as pd

import requests

pd.set_option('display.max_columns', None)
```

```python
def url_dinamica (genero, anio, tipo, clave):

    lista_pelis = []
    
    for pag in range (1,51):

        url = "https://moviesdatabase.p.rapidapi.com/titles"
        querystring = {"genre":genero,"titleType":tipo,"year":anio,"page":pag}
        headers = {
        "X-RapidAPI-Key": clave,
        "X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"}

        response = requests.get(url, headers=headers, params=querystring)

        response_json = response.json()

        lista_pelis.append (response_json)
       

    return lista_pelis
```

```python
#prueba rango años:
lista_jsons = []
for anio in range(2000,2024):

    movie_drama = url_dinamica("Drama", anio, "movie","54b0a47db5mshf6ed07e25abb30dp19e0edjsn935796beb60a")
    lista_jsons.append(movie_drama)
```

```python
def busqueda_info (lista_json):
    lista_final= []
    for num in range (1,len(lista_json)):
        for pelicula in lista_json[num]["results"]:
            lista_para_tupla = []
            lista_para_tupla.append(pelicula["titleType"]["text"])
            lista_para_tupla.append(pelicula['originalTitleText']['text'])
            try:      
                lista_para_tupla.append(pelicula["releaseDate"]["year"])
            except:
                lista_para_tupla.append(pelicula["releaseYear"]["year"])
            try:
                lista_para_tupla.append(pelicula['releaseDate']['month'])
            except:
                lista_para_tupla.append(0)
            lista_para_tupla.append(pelicula['id'])
            tupla = tuple(lista_para_tupla)
            lista_final.append(tupla)
                
    return lista_final
```

```python
#fin = len(lista_jsons)
#print(fin)

lista_total = []

for num in range(0, 9):
    lista_movie_2015_drama = busqueda_info(lista_jsons[num])
    lista_total = lista_total + lista_movie_2015_drama

lista_total
```
