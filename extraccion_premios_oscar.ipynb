```python
# antes de empezar importamos las librerías que vamos a usar. 
# Importar librerías para web scraping y manipulación de datos
# -----------------------------------------------------------------------
from bs4 import BeautifulSoup
import requests

# Importar librerías para manipulación y análisis de datos
# -----------------------------------------------------------------------
import pandas as pd

# Importar librerías para procesamiento de texto
# -----------------------------------------------------------------------
import re
```

```python
# al igual que en el ejemplo anterior lo primero que haremos será definir la url de la página de la vamos a sacar datos
url_oscar = "https://es.wikipedia.org/wiki/Premios_%C3%93scar#Categor%C3%ADas"

# hacemos la request a la página de la que queremos sacar la info
res_oscar = requests.get(url_oscar)

# vemos si todo ha ido bien
print("La respuesta de la petición es:", res_oscar.status_code)
```

```python
# creamos el objeto BeautifulSoup para poder acceder al contenido solicitado
sopa_oscar = BeautifulSoup(res_oscar.content, 'html.parser')

# recordemos que el método .prettify nos permite mostrar de una forma más amigable los resultados obtenidos de la sopa
print(sopa_oscar.prettify())
```

```python
# vamos a seguir usando el metodo ".find_all()", pero en este caso lo que buscaremos son todas las tablas que tenemos en la página web.
tablas = sopa_oscar.find_all("table")

print("El número de tablas que tenemos en la página web es:", len(tablas))
```

```python
# veamos que tenemos en cada una de las tablas que hemos extraído de la página web. 
print("En la tabla 1 tenemos: \n ", tablas[0])
print("\n---------------------\n")

print("En la tabla 2 tenemos: \n", tablas[1])
print("\n---------------------\n")

print("En la tabla 3 tenemos: \n", tablas[2])

print("En la tabla 4 tenemos: \n ", tablas[3])
print("\n---------------------\n")
```

```python
#la tabla que queremos es la tabla 3, cuyo indice es el 2.
```

```python
# La tabla que nos interesa es la última por lo que vamos a crear una nueva variable donde almacenemos los resultados de la tabla que nos interesa
nuestra_tabla = tablas[2]

# lo primero que nos va a interesar es extraer los encabezados de la tabla, teniendo en cuenta las definiciones que hemos visto antes, 
# seleccionaremos la etiqueta "th" usando el método "th"
lista_encabezados = nuestra_tabla.find_all("th")

print("La lista que nos devuelve el metodo '.find_all()' es:\n", lista_encabezados)

# como hemos estado haciendo hasta ahora, tendremos que iterar por la lista obtenida en el paso anterior y extraer el texto de cada elemento
encabezados_oscar = [columna.text for columna in lista_encabezados]

print("\n-------------------------\n")
print("Los encabezados de la tabla son:", encabezados_oscar)
```

```python
encabezados_premios = encabezados_oscar [1:6]
encabezados_premios
```

```python
# Eliminar el carácter \n de cada elemento en la lista
lista_encabezados = [elemento.strip() for elemento in encabezados_premios]

print(lista_encabezados)
```

```python
# sacamos todas las filas de la variable que nos hemos creado previamente, usando el método .find_all() y descartamos las que no nos interesan
filas_oscar = nuestra_tabla.find_all("tr")

print("La cantidad de filas que hemos extraido de la tabla es:", len(filas_oscar))

print("\n----------------------\n")
print("El contenido que tenemos en la primera fila es:\n", filas_oscar[72] )

print("\n----------------------\n")
print("El contenido que tenemos en la segunda fila es:\n", filas_oscar[95] )
```

```python
# Creamos una lista llamada 'resultados_oscar' para almacenar los resultados obtenidos.
resultados_oscar = []

# Iniciamos un bucle 'for' para iterar a través de la lista 'filas_oscar', pero comenzamos desde el elemento 72 de la lista.
for fila in filas_oscar[72:]:
    # Para cada 'fila', extraemos el texto, lo dividimos en una lista usando '\n' como separador y eliminamos el primer y último elemento de la lista.
    fila_texto = fila.text
    elementos_fila = fila_texto.split("\n")

    # Añadimos la lista de elementos a la lista 'resultados_oscar'.
    resultados = []
    resultados.append(elementos_fila[3])
    resultados.append(elementos_fila [5])
    resultados.append(elementos_fila [7])
    resultados.append(elementos_fila [9])
    resultados.append(elementos_fila [11])
    resultados = tuple (resultados)
    resultados_oscar.append (resultados)


# Imprimimos los resultados obtenidos después de iterar por la lista.
print("Los resultados de iterar por la lista son:\n", resultados_oscar)
```

```python
df_oscar = pd.DataFrame(resultados_oscar)
df_oscar
```

```python
df_oscar.to_csv("premios_oscar.csv")
```
